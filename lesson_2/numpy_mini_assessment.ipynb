12. Create an array of 360 random numbers from -10 to 50 (to resemble the weather) and find the min, max, 25th percentile, 75 percentile, and the median.

```python
weather = np.random.randint(low=-10, high=50, size=360, dtype='int32')
print(weather)
print(f"Minimum Value: {np.min(weather)}")
print(f"Maximum Value: {np.max(weather)}")
print(f"25th percentile: {np.percentile(weather, 25)}")
print(f"Median: {np.median(weather)}")
print(f"75th percentile: {np.percentile(weather, 75)}")
```

13. Reshape the array you created in step 12 into a matrix of 300 by 12 ad get the same values but only for the months this time.

```python
weatherM = weather.reshape(30,12)
print(weatherM)
print(f"Minimum Value: {np.min(weatherM,axis=0)}")
print(f"Maximum Value: {np.max(weatherM,axis=0)}")
print(f"25th percentile: {np.percentile(weatherM,25,axis=0)}")
print(f"Median: {np.median(weatherM,axis=0)}")
print(f"75th percentile: {np.percentile(weatherM,75,axis=0)}")
```

14. Create a mask for your data, and filter out and print:
- average temperature of the days where the temperature was below 10
- average temperature of the days where the temperature was greater than 30

```python
bad_days_mask = (weatherM < 10)
bad_days_mask
bad_days = weatherM[bad_days_mask]
bad_days
print("The Average temperature on bad days was %i F ☀️" % np.mean(bad_days))
```

```python
nice_days_mask = (weatherM > 10)
nice_days_mask
nice_days = weatherM[nice_days_mask]
nice_days
print("The Average temperature on nice days was %i F ☀️" % np.mean(nice_days))
```
