```python
#Create a new column Category where movies release in 2019 are categorized as 'New Release' 
#and movies released in 2016,2017,2018 are categorized as 'A Bit old' and anything older than that is categorized 'Old but Gold'. 
#List random 10 records from the dataframe and verify the same.

# Create a list to store the data
Category = []


for i in netflix_df['release_year']:
    if i >= 2019:
        Category.append('New Release')
    elif i >= 2016:
        Category.append('A Bit old')
    else:
        Category.append('Old but Gold')
        
# Create a column from the list
netflix_df['Category'] = Category

netflix_df.head()

#ten random records
random_subset = netflix_df.sample(n=10)
random_subset.head()
```

```python
netflix_df.drop(['Category'], axis=1, inplace=True)
netflix_df.head()
```
