9. Create a new column `Category` where movies release in 2019 are categorized as 'New Release' and movies released in 2016,2017,2018 are categorized as 'A Bit old' and anything older than that is categorized 'Old but Gold'. List random 10 records from the dataframe and verify the same.

```python
netflix_df.loc[:,'Category'] = netflix_df['release_year'].apply(lambda x: 'New Release' if x == 2019  else ('A Bit Old' if (x <= 2018 and x > 2015) else 'Old but Gold'))
netflix_df.head(10)
```
